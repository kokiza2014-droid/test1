```python
#check point jeu de devinette
import random

x = random.randint(1,100)
while True:
    z=int(input("vous etes inviter à entrer votre réponse sous forme de chaîne de caractères"))
   
    if z==x:
        print(f"Félicitations ! Vous avez deviné le nombre correctement! qui est {z}")
        break
    elif z>x:
        print("Votre réponse est trop élevée. Réessayez")
       
        
    else: 
        print("Votre réponse est trop basse. Réessayez")
```
